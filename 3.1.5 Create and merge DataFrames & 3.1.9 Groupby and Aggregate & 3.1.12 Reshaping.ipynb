# 3.1.9 Practical activity: Use groupby() and aggregate() functions

Use the groupby() and aggregate() functions on the merged DataFrame to:

Determine how many films from each year (released from 2012 to the present) were watched on Netflix.
Determine what the average runtime is of movies released each year.
Determine what the best and worst reviews are that movies received on Rotten Tomatoes.
Approach

```python
# Determine how many films from each year (released from 2012 to the present) were watched on Netflix.

movies_final_SUM = movies_final.groupby('Year')[['Netflix']] \
.sum().reset_index() \
.sort_values('Netflix', ascending=False)
movies_final_SUM[movies_final_SUM['Year'] >= 2012]
```

```python
 # Determine what the average runtime is of movies released each year.
movies_final_AVG = movies_final.groupby('Year')[['Runtime']] \
.mean().reset_index() \
.sort_values('Runtime', ascending=False)  
movies_final_AVG[movies_final_AVG['Year'] >= 2012]
```

```python
# Determine the number of movies watched on Netflix since 2012.
mo_gpby = movies_final.groupby(['Year'])['Netflix'].agg('sum').reset_index()
mo_gpby[mo_gpby['Year'] >= 2012]
```

```python
# Determine what the best and worst reviews are that movies received on Rotten Tomatoes.

movies_final.sort_values('Rotten Tomatoes', ascending=False)
```

```python
movies_final.sort_values('Rotten Tomatoes', ascending=True)
```

```python
movies_final_MINMAX = movies_final.groupby('Year')[['Rotten Tomatoes']] \
.agg(['min','max']).reset_index()
movies_final_MINMAX[movies_final_MINMAX['Year'] >= 2012]
```
