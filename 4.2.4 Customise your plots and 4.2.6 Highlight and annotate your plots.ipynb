#  4.2.4 Practical activity Customise your plots

```python
# import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Read the CSV file.
ott_merge = pd.read_excel('ott.xlsx')

# Determine shape of classlist.
print(ott_merge.shape)
ott_merge.head()
```

```python
# Read the Excel file.
movies_merge = pd.read_csv('movies.csv')

# Determine shape of classlist.
print(movies_merge.shape)
movies_merge.head()
```

```python
movies_final = movies_merge.merge(ott_merge, how='left', on='ID')
movies_final.head()
```

```python
mov_ott_2010 = movies_final[movies_final['Year']>=2010]

ax = sns.countplot(x='Year', data=mov_ott_2010)

ax.set(ylabel='Percent')

total = len(mov_ott_2010['Year'])

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x()
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y))

plt.xticks(rotation=90)
plt.show()
```

```python
# Plot a histogram for the IMDb rating column.
sns.histplot(data=movies_final, x='IMDb', binwidth=1)
```

```python
ax = sns.displot(data=movies_final, x='IMDb', bins=10,kind='hist', 
                 palette='GnBu', aspect=1.4, kde=True)

plt.show()
```

# 4.2.6 Practical activity: Highlight and annotate your plots

```python
# Create a scatterplot with two variables (IMDb and Rotten Tomatoes).
sns.scatterplot(x='IMDb', y='Rotten Tomatoes', hue='Netflix', data=movies_final)
```

```python
sns.boxplot(data=movies_final, 
            x='Age', 
            y='IMDb', 
            hue='Netflix',
            palette='inferno',
            linewidth=2,
            showfliers = False)

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
sns.lineplot(x = 'Year', y = 'IMDb', data=movies_final, linewidth=5, errorbar=('ci', 0))
```

```python
sns.lineplot(x = 'Year', y = 'IMDb',
             data=movies_final[movies_final['Age'].isin(['16+', '18+'])], 
             hue ='Age')
```

```python
sns.lineplot(x = 'Year', y = 'IMDb', 
             data=movies_final[movies_final['Age'].isin(['16+', '18+'])], 
             hue ='Age', style = 'Age', markers=True, ci=0)
```
