```python
# Import libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create an empty plot.
fig, ax = plt.subplots()

# Create variables.
x = np.linspace(0, 10, 10000)
y = np.exp(x)

# Create plot.
ax.plot(x, y)
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Create variables.
x = np.linspace(0, 10)
y = np.exp(x)

# Set labels and title.
ax.set_xlabel("Time Elapsed (minutes)")
ax.set_ylabel("Count")
ax.set_title("Bacteria Growth")

# Create plot.
ax.plot(x, y)
```

```python
# Add ticks to x-axis.
ax.set_xticks([0, 5, 10])
ax.set_xticks([1, 2, 3, 4, 6, 7, 8, 9], minor=True)

# View figure.
fig
```

```python
# Add ticks labels to x-axis.
ax.set_xticklabels(['0.0', '5.0', '10.0'], minor=False)

# View figure.
fig
```

```python
# Add ticks labels to x-axis.
ax.set_xticklabels(['1', '2', '3', '4', '6',
                    '7', '8', '9'], minor=True)

# View figure.
fig
```

```python
# Create new variables.
y_no_food = np.exp(x)
y_food = np.exp(x + 1.01)

# Create an empty plot.
fig, ax = plt.subplots()

# Plot the new variables and add the label to the legend.
lines = ax.plot(x, y_no_food)
line = lines[0]

# Set the line variable.
line.set_label("No Food")

# Add legend.
ax.legend()
```

```python
fig
```

```python
# Plot second line.
ax.plot(x, y_food, label="With Food")

# Create legend.
ax.legend()

# Specify legend position.
ax.legend(loc='center left')

# View figure.
fig
```
