```python
# Import the necessary libraries.
import pandas as pd

# Import the Excel file.
transport = pd.read_excel('transport_costs.xlsx')

# View the DataFrame.
print(transport.shape)
print(transport.dtypes)
transport
```

```python
# Drop columns currency, distance unit.
transport_subset = transport[['region','country','port','sea freight cost','road transport cost per km']]

# view the dataset.
transport_subset
```

```python
# Create new columns with calculated values.
transport_subset['total_sea_freight_costs'] = transport_subset['sea freight cost'] * 2 * 4
transport_subset['total_road_transport_costs'] = transport_subset['road transport cost per km'] * 10000
transport_subset['total_export_costs'] = transport_subset['total_sea_freight_costs'] 
+ transport_subset['total_road_transport_costs']

# View output.
transport_subset
```

```python
transport_reorder = transport_subset[['region','country','port','sea freight cost','total_sea_freight_costs',
                               'road transport cost per km','total_road_transport_costs',
                              'total_export_costs']]
transport_reorder
```
